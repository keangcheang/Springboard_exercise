```python
#remake the data to daily data and visualize
login_df = pd.read_json("logins.json")
login_df.set_index('login_time', inplace=True)
```

```python
jan_df = login_df['1970-01'].reset_index()
feb_df = login_df['1970-02'].reset_index()
mar_df = login_df['1970-03'].reset_index()
apr_df = login_df['1970-04'].reset_index()
```

```python
month_list = [jan_df, feb_df, mar_df, apr_df]

for month_df in month_list: 
    month_df['dayofweek'] = month_df['login_time'].map(lambda x: x.weekday())
    
print jan_df.head(3)
```

```python
def color_plot(df, title, weekends=False): 
    new_df = df.set_index('login_time')

    plt.figure(figsize=[20,5])
    for day in sorted(list(new_df['dayofweek'].unique())):

        count = new_df[new_df['dayofweek'] == day]
        count.drop('dayofweek', axis=1, inplace=True)
        count['count'] = 1
        count = count.resample('15T').sum()
    
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=15)
        if weekends == True:
            if day in [0,1,2,3,4]:
                color = 'red'
            else:
                color = 'black'
            plt.plot(count, color=color)
         
        else:
            plt.plot(count)
        plt.title(title, fontsize=20)
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
color_plot(jan_df, "January Plot of Daily Logins")
```

```python
color_plot(jan_df, "January Plot of Weekend Logins", weekends=True)
```

```python
color_plot(feb_df, "February Plot of Daily Logins")
```

```python
color_plot(feb_df, "February Plot of Weekend Logins", weekends=True)
```

```python
color_plot(mar_df, "March Plot of Daily Logins")
```

```python
color_plot(mar_df, "March Plot of Weekend Logins", weekends=True)
```

```python
color_plot(apr_df, "April Plot of Daily Logins")
```

```python
color_plot(apr_df, "April Plot of Weekend Logins", weekends=True)
```

As mentioned earlier, weekends is a little bit busier than weekedays and busier at night. Weekday has the highest demands around noon. Also, keeping in mind that April logins has the date span to the 14th. 

## Part 2: Experiment and Metrics Design

1. What would you choose as the key measure of success of this experiment in encouraging driver partners to serve both cities, and why?
   The key measure of success for this challenge can be the measure of the mean in gross profit between before and after the new implementation which is the all toll cost reimbursing to all the drivers. Another measure would be to compare the activites of each transport in each city changes before and after the new implementation. In other word, to check the distribution disperse to other cities or not. 

2. Describe a practical experiment you would esign to compare the ffectiveness of the proposed change in relation to the key measure of success.
   In order to implement the key measure above, one would need to do some statistical testing on the mean of the gross profit for pair observation. 
   The statistical test: 
       - null hypothesis: the mean of gross profit of Ultimate Gotham and Metropolis between no toll-cost reimbursment and toll-cost reimbursment are the same
       - alternative hypothesis: the mean of gross profit of Ultimate Gotham and Metropolis between no toll-cost reimbursment and toll-cost reimbursment are not the same
       
     If the p-value is below 5% significant level then we reject the null hypothesis and we believe that reimbursing the toll cost to all the drivers would help improve the business and increase performance of transport.

```python
#setting the date format right
ultimate_df['last_trip_date'] = pd.to_datetime(ultimate_df['last_trip_date'])
ultimate_df['signup_date'] = pd.to_datetime(ultimate_df['signup_date'])
ultimate_df['last_trip_date'].head(3)
```

```python
#create a retained columns
#the last date of the data set is 07/01/2014 then the retained riders will be the last 30 day of July 1st. 
ultimate_df['retained'] =np.nan
ultimate_df.loc[ultimate_df['last_trip_date'] < '2014-06-01',"retained"] =0
ultimate_df.loc[ultimate_df['last_trip_date'] >= '2014-06-01', "retained"] =1
```

```python
ultimate_df.head(3)
```

```python
#model building
from sklearn.metrics import roc_curve, auc, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
#
final_ult_df = pd.get_dummies(ultimate_df,columns=['city', 'ultimate_black_user', 'phone'] )
final_ult_df.head(3)
```

```python
final_ult_df= final_ult_df.drop(['last_trip_date', 'signup_date'], axis=1)
```

```python
#split the dataset into train and test
train, test = train_test_split(final_ult_df, test_size = .30, stratify=final_ult_df['retained'])
train.reset_index(drop=True, inplace=True)
test.reset_index(drop=True, inplace=True)
print train.shape
print test.shape
```

```python
target = train['retained']
train = train.drop('retained', axis=1)
target_test = test['retained']
test = test.drop('retained', axis=1)
```

```python
clf = KNeighborsClassifier()
fit_knn=clf.fit(train, target)
fit_knn.predict(test)
print ("Accurary Score: {}").format(clf.score(test, target_test))
```

```python
clf = DecisionTreeClassifier()
fit_tree= clf.fit(train, target)
fit_tree.predict(test)
print ("Accurary Score: %.4f")%clf.score(test, target_test)
```

```python
feature_importance = pd.DataFrame(fit_tree.feature_importances_).T
feature_importance.columns = list(train.columns)
feature_importance
```

### The KNN seems to perform a little bit better than decision tree algorithm. 
### Based on the result above, the four important features predicting the retained riders are (in order from strong to weak): avg_dist, avg_rating_by_driver, weekday_pct, avg_rating_of_driver. Therefore, in order to increase the riders, Ultimate should improve on this main features. The model can be improved if one use cross validation and hyperparameters tuning.
