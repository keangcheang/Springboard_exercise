```python
#remake the data to daily data and visualize
login_df = pd.read_json("logins.json")
login_df.set_index('login_time', inplace=True)
```

```python
jan_df = login_df['1970-01'].reset_index()
feb_df = login_df['1970-02'].reset_index()
mar_df = login_df['1970-03'].reset_index()
apr_df = login_df['1970-04'].reset_index()
```

```python
month_list = [jan_df, feb_df, mar_df, apr_df]

for month_df in month_list: 
    month_df['dayofweek'] = month_df['login_time'].map(lambda x: x.weekday())
    
print jan_df.head(3)
```

```python
def color_plot(df, title, weekends=False): 
    new_df = df.set_index('login_time')

    plt.figure(figsize=[20,5])
    for day in sorted(list(new_df['dayofweek'].unique())):

        count = new_df[new_df['dayofweek'] == day]
        count.drop('dayofweek', axis=1, inplace=True)
        count['count'] = 1
        count = count.resample('15T').sum()
    
        plt.xticks(fontsize=10)
        plt.yticks(fontsize=15)
        if weekends == True:
            if day in [0,1,2,3,4]:
                color = 'red'
            else:
                color = 'black'
            plt.plot(count, color=color)
         
        else:
            plt.plot(count)
        plt.title(title, fontsize=20)
```

```python
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
color_plot(jan_df, "January Plot of Daily Logins")
```

```python
color_plot(jan_df, "January Plot of Weekend Logins", weekends=True)
```

```python
color_plot(feb_df, "February Plot of Daily Logins")
```

```python
color_plot(feb_df, "February Plot of Weekend Logins", weekends=True)
```

```python
color_plot(mar_df, "March Plot of Daily Logins")
```

```python
color_plot(mar_df, "March Plot of Weekend Logins", weekends=True)
```

```python
color_plot(apr_df, "April Plot of Daily Logins")
```

```python
color_plot(apr_df, "April Plot of Weekend Logins", weekends=True)
```

As mentioned earlier, weekends is a little bit busier than weekedays and busier at night. Weekday has the highest demands around noon. Also, keeping in mind that April logins has the date span to the 14th.

```python
import json
import seaborn as sn
```

```python
file = open('ultimate_data_challenge.json', 'r')
ultimate_df = pd.DataFrame(json.load(file))
file.close()

ultimate_df.head(3)
```

```python
ultimate_df.info()
#there are three attributes that have missing value: avg_rating_by_driver, avg_rating_of_driver and phone. 
```

```python
ultimate_df.describe()
```

```python
#missing values
print ("missing value: {} in avg_rating_by_driver").format(ultimate_df.avg_rating_by_driver.isnull().sum())
print ("missing value: {} in avg_rating_of_driver").format(ultimate_df.avg_rating_of_driver.isnull().sum())
print ("missing value: {} in phone").format(ultimate_df.phone.isnull().sum())
```

```python
phone= list(ultimate_df.groupby('phone').size())
plt.bar([1,2], phone)
plt.xticks([1,2],["android", 'iphone'])
plt.title("Phone");
```

```python
plt.hist(ultimate_df[ultimate_df.avg_rating_by_driver.notnull()]["avg_rating_by_driver"])
plt.title("Histogram: avg_rating_by_driver");
```

```python
plt.hist(ultimate_df[ultimate_df.avg_rating_of_driver.notnull()]["avg_rating_of_driver"])
plt.title("Histogram: avg_rating_of_driver");
```

```python
ultimate_df['avg_rating_by_driver'].fillna(ultimate_df['avg_rating_by_driver'].mean(), inplace=True)
ultimate_df['avg_rating_of_driver'].fillna(ultimate_df['avg_rating_of_driver'].mean(), inplace=True)
ultimate_df['phone'].fillna(method='ffill', inplace=True)
```

```python
ultimate_df.info()
# no more missing values
```

```python
plt.figure(figsize=[15,10])
sn.boxplot(ultimate_df);
```
